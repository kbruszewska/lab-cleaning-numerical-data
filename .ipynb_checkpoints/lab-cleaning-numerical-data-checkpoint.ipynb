#### 7. Define a function that differentiates between continuous and discrete variables. Hint: The number of unique values might be useful. Store continuous data into a continuous_df variable and do the same for discrete_df

General ideas, how to share features into:
- Continuous: float type, many different values (unique number of values above 59). can be also integer, when we have many unique values within the dataset. 
- Discrete: int type, only few values, also datetype


```python
def continuous_or_discrete(df):
    # df Input dataframe
    # Output continuous_df for dataframe with continuous values and discrete_df for dataframe with discrete values
    continuous_df = pd.DataFrame()
    discrete_df = pd.DataFrame()
    column_names = df.columns
    for col in column_names:
        if df[col].dtype == 'float64':
            continuous_df = pd.concat([continuous_df, df[col]], axis=1)
        elif df[col].dtype == 'datetime64[ns]':
            discrete_df = pd.concat([discrete_df, df[col]], axis=1)
        elif (df[col].dtype == 'int64') & (df[col].nunique() < 60):
            discrete_df = pd.concat([discrete_df, df[col]], axis=1)
        else:
            continuous_df = pd.concat([continuous_df, df[col]], axis=1)
    return continuous_df, discrete_df
```

```python
continuous_df, discrete_df = continuous_or_discrete(numerical_df)
```

```python
def plot_continuous_variable(df):
    # df Input is a dataframe with continuous variables
    # Output should be distplot for each variable.
    for col in df:
        sns.distplot(df[col])
        plt.title(f'Distribution Plot for {col}')
        plt.show()
```

```python
plot_continuous_variable(continuous_df)
```

```python
def plot_discrete_variable(df):
    # df Input is a dataframe with discrete variables
    # Output should be histogram for each variable.
    for col in df:
        df[col].hist(bins=100)
        plt.title(f'Distribution Plot for {col}')
        plt.show()
```

```python
plot_discrete_variable(discrete_df)
```
